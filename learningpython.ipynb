## 1. Variables & Data Types
Mencoba berbagai tipe data dasar Python.


```python
name = "Aqis"
age = 19
height = 158.5
is_student = True

print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Student:", is_student)
print("Type of age:", type(age))
```

## 2. Operators
Simulasi perhitungan sederhana


```python
price = 12000
quantity = 3
total = price * quantity

discount = 0.1
final_price = total - (total * discount)

print("Total:", total)
print("After discount:", final_price)
```

## 3. Conditional (if-else)
Menentukan kategori nilai mahasiswa


```python
score = 82

if score >= 85:
    grade = "A"
elif score >= 70:
    grade = "B"
elif score >= 60:
    grade = "C"
else:
    grade = "D"

print("Score:", score)
print("Grade:", grade)
```

## 4. Loops
Menampilkan angka genap dari 1-20


```python
for i in range(1, 21):
    if i % 2 == 0:
        print(i, end=" ")
```

## 5. List & Dictionary
Menyimpan data mahasiswa


```python
students = ["Aqis", "Rina", "Dika"]

for s in students:
    print("Student:", s)

student_data = {
    "name": "Aqis",
    "major": "Informatics",
    "semester": 3
}

for key, value in student_data.items():
    print(key, ":", value)
```

## 6. Functions
Membuat fungsi untuk menghitung luas lingkaran


```python
def circle_area(radius):
    pi = 3.14159
    return pi * radius ** 2

print("Area:", circle_area(7))
```

## 7. External Libraries
Menggunakan library numpy


```python
import numpy as np

data = np.array([72, 80, 65, 90, 88])

print("Mean:", np.mean(data))
print("Max:", np.max(data))
print("Min:", np.min(data))
```
